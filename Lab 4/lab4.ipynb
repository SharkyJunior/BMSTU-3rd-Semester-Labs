```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('gdp.csv').dropna()
df.head()
```

```python
region = 'Germany'

subdf = df[df['Country Name'] == region]

reg = LinearRegression().fit(subdf[['Year']], subdf['Value'] / 10**9)

plt.figure(figsize=(10, 8))

plt.scatter(subdf['Year'], subdf['Value'] / 10**9)
plt.plot(subdf['Year'], reg.predict(subdf[['Year']]), color='red')
plt.xlabel('Year')
plt.ylabel('GDP (billion USD)')
plt.show()
```

```python
reg.score(subdf[['Year']], subdf['Value'] / 10**9)
```

```python
year_to_predict = 2026

print(f'{reg.predict([[year_to_predict]])} billion USD')
```
