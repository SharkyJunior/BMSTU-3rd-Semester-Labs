```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df3.csv', sep='    ', names=['x', 'y'])
df.head()
```

```python
df.plot.scatter(x='x', y='y')
```

```python
kmeans = sk.cluster.KMeans(n_clusters=15).fit(df.to_numpy())
labels = kmeans.labels_

plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], c=labels)
plt.show()
```

```python
agglomerative = sk.cluster.AgglomerativeClustering(n_clusters=15).fit(df.to_numpy())
labels = agglomerative.labels_

plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], c=labels)
plt.show()
```

```python
dbscan = sk.cluster.DBSCAN(eps=20000, min_samples=4).fit(df.to_numpy())
labels = dbscan.labels_

plt.figure(figsize=(10, 10))
plt.scatter(df['x'], df['y'], c=labels, cmap='rainbow')
plt.show()
```
